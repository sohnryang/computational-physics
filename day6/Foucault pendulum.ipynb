# 푸코 진자

다음과 같은 식을 따른다. <strike>증명은 여백이 부족하므로 생략</strike>

$$
\ddot{x'} = \frac{F_x}{m} - 2\omega(\dot{z'} \cos{\lambda} - \dot{y'} \sin{\lambda})
$$

$$
\ddot{y'} = -\omega^{2}y - 2\omega(\dot{x'} \sin{\lambda})
$$

import를 하자.

```python
import numpy as np
import matplotlib.pyplot as plt
from differential_equation import rk2_algorithm
%matplotlib inline
```

초기 조건을 정하자.

```python
init = np.array([0.01, 0, 0.01, 0.1])
w = 1
W = 0.2
c = np.pi / 6

def foucalt(status, t):
    v_x = status[1]
    a_x = -(w ** 2) * status[0] + 2 * W * status[3] * np.sin(c)
    v_y = status[3]
    a_y = -(w ** 2) * status[2] - 2 * W * status[1] * np.sin(c)
    return np.array([v_x, a_x, v_y, a_y])
```

이제 미분방정식을 풀어보자.

```python
solution = rk2_algorithm(foucalt, 0, 100, init, 1000)
```

그래프로 그려보자.

```python
plt.figure(figsize=(8, 8))
plt.plot(solution[0][:,0], solution[0][:,2], '-')
plt.show()
```
